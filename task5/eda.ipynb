```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Titanic.csv")

# Display first five rows
df.head()
```

```python
# Dataset Information
df.info()
```

```python
# Statistical Summary
df.describe()
```

```python
# Count of passengers by gender
df['Sex'].value_counts()
```

```python
# Histogram of Age

plt.figure(figsize=(8,5))
plt.hist(df['Age'].dropna(), bins=20, edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# Boxplot of Fare

plt.figure(figsize=(8,5))
sns.boxplot(x=df['Fare'])
plt.title('Boxplot of Fare')
plt.show()
```

```python
# Scatter Plot: Age vs Fare

plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Correlation Heatmap

plt.figure(figsize=(8,6))

numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairplot

sns.pairplot(
    df[['Survived', 'Pclass', 'Age', 'Fare']].dropna(),
    hue='Survived'
)

plt.show()
```

# Observations

## Dataset Information
- The dataset contains 891 passenger records and 12 columns.
- Missing values are present in the Age, Cabin, and Embarked columns.

## Statistical Summary
- The average passenger age is around 29.7 years.
- Passenger fares vary widely, indicating different ticket classes.

## Gender Distribution
- The dataset contains more male passengers than female passengers.

## Boxplot
